```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,11)
```

```python
y=x*2
```

```python
plt.plot(x,y)
plt.show()
```

<!-- giving colour , linestyle and width to line -->

```python
plt.plot(x,y, color='y', linestyle='--', linewidth='3')
plt.show()
```

```python
plt.plot(x,y, color='y', linestyle='--', linewidth='3')
plt.title('x vs y')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
